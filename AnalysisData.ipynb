# Experiment 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from functools import reduce
import matplotlib as mpl
```

```python
df = pd.read_csv('test_disparity_gis8 experiment-traffic-flood-3-table2.csv')
```

```python
for i in range(0,11, 1):
    df[df['flood-level'] == (i/10)].boxplot(column=['percent-different'], by=['traffic-level'])
    plt.suptitle('')
    plt.title("flood-level " + str(i/10)).set_color('cyan')
    plt.xlabel("traffic-level").set_color('cyan')
    plt.ylabel("Same Neighbors Percentage").set_color('cyan')
    plt.tick_params(axis='x', colors='cyan')
    plt.tick_params(axis='y', colors='cyan')
    plt.tight_layout()
    plt.show()
```

```python
for i in range(0,11, 1):
    df[df['flood-level'] == (i/10)].boxplot(column=['count poor with [is_city = 1]'], by=['traffic-level'])
    plt.suptitle('')
    plt.title("flood-level " + str(i/10)).set_color('black')
    plt.xlabel("traffic-level").set_color('black')
    plt.ylabel("Poor Inside JKT").set_color('black')
    plt.tick_params(axis='x', colors='black')
    plt.tick_params(axis='y', colors='black')
    plt.tight_layout()
    plt.show()
```

```python
for i in range(0,11, 1):
    df[df['flood-level'] == (i/10)].boxplot(column=['mean [home_quality] of poor'], by=['traffic-level'])
    plt.suptitle('')
    plt.title("flood-level " + str(i/10)).set_color('black')
    plt.xlabel("traffic-level").set_color('black')
    plt.ylabel("home price of poor").set_color('black')
    plt.tick_params(axis='x', colors='black')
    plt.tick_params(axis='y', colors='black')
    plt.tight_layout()
    plt.show()
```

```python
for i in range(0,11, 1):
    df[df['flood-level'] == (i/10)].boxplot(column=['median [ min [distance myself] of jobs ] of rich'], by=['traffic-level'])
    plt.suptitle('')
    plt.title("flood-level " + str(i/10)).set_color('black')
    plt.xlabel("traffic-level").set_color('black')
    plt.ylabel("Distance to Work of the Rich").set_color('black')
    plt.tick_params(axis='x', colors='black')
    plt.tick_params(axis='y', colors='black')
    plt.tight_layout()
    plt.show()
```

```python
for i in range(0,11, 1):
    df[df['flood-level'] == (i/10)].boxplot(column=['median [ min [distance myself] of jobs ] of poor'], by=['traffic-level'])
    plt.suptitle('')
    plt.title("flood-level " + str(i/10)).set_color('black')
    plt.xlabel("traffic-level").set_color('black')
    plt.ylabel("Distance to Work of the Poor").set_color('black')
    plt.tick_params(axis='x', colors='black')
    plt.tick_params(axis='y', colors='black')
    plt.tight_layout()
    plt.show()
```

```python
for i in range(0,11, 1):
    df[df['flood-level'] == (i/10)].boxplot(column=['mean [home_price] of rich'], by=['traffic-level'])
    plt.suptitle('')
    plt.title("flood-level " + str(i/10)).set_color('black')
    plt.xlabel("traffic-level").set_color('black')
    plt.ylabel("Home Price of the Rich").set_color('black')
    plt.tick_params(axis='x', colors='black')
    plt.tick_params(axis='y', colors='black')
    plt.tight_layout()
    plt.show()
```

# Experiment 2

```python
df = pd.read_csv('test_disparity_gis8 experiment-3-table6.csv')
df.head()
```
